```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_data = pd.read_csv('Resources/got.csv')
got_data
```

```python
# Get attacker and defender king data
attacker_data = got_data['attacker_king'].value_counts()
defender_data = got_data['defender_king'].value_counts()
attacker_data
```

```python
# Get total battle data
total_battle = attacker_data.add(defender_data, fill_value=0)
# total_battle = total_battle.add(attacker_data, fill_value=0)
```

```python
# Configure plot and ticks
total_battle.plot(kind='bar', facecolor='red')

# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
plt.xlabel("King")
plt.xticks(rotation=45)

# Show plot
plt.show()

# Resize plot to display labels
plt.tight_layout()
```
